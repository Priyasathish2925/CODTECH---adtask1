```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("Big Data Analysis with PySpark") \
    .getOrCreate()
```

```python
data = spark.read.csv("task1.csv", header=True, inferSchema=True)
```

```python
data.printSchema()
data.show(5)
```

```python
cleaned_data = data.dropna()
```

```python
grouped_data = cleaned_data.groupBy("Category").count()
grouped_data.show()
```

```python
average_data = cleaned_data.groupBy("Category").avg("Sales")
average_data.show()
```

```python
sorted_data = cleaned_data.orderBy("Sales", ascending=False)
sorted_data.show(10)
```
